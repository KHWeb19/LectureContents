```python
# csv -> DB
db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = mysql_user_info['user_id'],
    passwd = mysql_user_info['password'],
    # 여기서 사용할 스키마 하나 생성!
    db = 'pydb'
)
```

```python
cursor = db.cursor()

sql = """
    create table csv_to_db_test(
        product_no int unsigned not null auto_increment,
        title varchar(200) not null,
        description text,
        price int not null,
        writer varchar(50) not null,
        reg_date timestamp default now(), 
        primary key(product_no)
    )
"""

cursor.execute(sql)
db.commit()
db.close()
print('테이블 생성 성공!')
```

```python
import csv
```

```python
f = open('db2csv.csv', 'r', encoding='UTF8')
#csv_reader = csv.reader(f, delimiter='\t')
csv_reader = csv.reader(f, delimiter=',')
# 구분자를 주의합시다 ;;;

print(csv_reader)
```

```python
cnt = 0

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
        
    print(row)
    product_no = (row[0])
    title = (row[1])
    description = (row[2])
    price = (row[3])
    writer = (row[4])
    reg_date = (row[5])
    
    print(product_no, title, description, price, writer, reg_date)
    
    sql = """
        insert into csv_to_db_test (title, description, price, writer, reg_date) values (%s, %s, %s, %s, %s)
    """
    
    cursor.execute(sql, (title, description, price, writer, reg_date))
    
db.commit()

f.close()
db.close()
```
